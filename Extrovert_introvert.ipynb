```python
sc_cols = ['Time_spent_Alone',  'Social_event_attendance', 'Going_outside', 'Friends_circle_size', 'Post_frequency']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
df1[sc_cols] = sc.fit_transform(df1[sc_cols])
```

```python
df1.head()
```

```python
X = df1[['Time_spent_Alone', 'Stage_fear', 'Social_event_attendance','Going_outside', 'Drained_after_socializing', 'Friends_circle_size', 'Post_frequency']]
```

```python
df1['Personality']=df1['Personality'].map({'Extrovert':1,'Introvert':0}).astype(int)
```

```python
Y = df1['Personality']
```

```python
Y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
x,y = smote.fit_resample(X,Y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_confusion_matrix
from mlxtend.classifier import StackingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# 20 % training dataset is considered for testing
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import StackingClassifier
#from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
```

```python
estimators = []
estimators.append(("KNearest Neighbours", KNeighborsClassifier()))
estimators.append(("Decision Tree", DecisionTreeClassifier()))
estimators.append(("Random Forest", RandomForestClassifier()))
estimators.append(("Gradient Boosting", GradientBoostingClassifier()))
```

```python
XGB = XGBClassifier()
```

```python
#Create a stacking classifier
from sklearn.ensemble import StackingClassifier
SC = StackingClassifier(estimators=estimators,final_estimator=XGB)
SC.fit(X_train, y_train)
y_pred = SC.predict(X_test)

print(f"\nStacking classifier training Accuracy: {SC.score(X_train, y_train):0.2f}")
print(f"Stacking classifier test Accuracy: {SC.score(X_test, y_test):0.2f}")
```
